# Q1: What is a random variable in probability theory
# A:
# A random variable is a function that assigns numerical values to outcomes of a random process or experiment.

# Q2: What are the types of random variables
# A:
# Two main types: discrete random variables (countable outcomes) and continuous random variables (uncountable outcomes over intervals).

# Q3: What is the difference between discrete and continuous distributions
# A:
# Discrete distributions describe probabilities of countable outcomes (e.g., number of heads). Continuous distributions describe probabilities over intervals (e.g., height).

# Q4: What are probability distribution functions (PDF)
# A:
# For continuous variables, the PDF describes the relative likelihood of outcomes. The area under the PDF curve over an interval gives the probability.

# Q5: How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)
# A:
# The CDF gives the probability that a variable is ≤ a value. It is the integral (sum) of the PDF up to that point.

# Q6: What is a discrete uniform distribution
# A:
# A distribution where all outcomes have equal probability. Example: rolling a fair die (each side has 1/6 chance).

# Q7: What are the key properties of a Bernoulli distribution
# A:
# Models a single trial with two outcomes: success (p) and failure (1-p). Mean = p, Variance = p(1-p).

# Q8: What is the binomial distribution, and how is it used in probability
# A:
# Models number of successes in n independent Bernoulli trials. Used for problems with fixed trials and constant success probability.

# Q9: What is the Poisson distribution and where is it applied
# A:
# Models the number of events in a fixed interval with known average rate (λ). Applied to rare events over time/space (e.g., calls to a hotline).

# Q10: What is a continuous uniform distribution
# A:
# A distribution where all values in an interval [a, b] are equally likely. PDF = 1/(b - a).

# Q11: What are the characteristics of a normal distribution
# A:
# Symmetric, bell-shaped, defined by mean (μ) and standard deviation (σ). Mean = median = mode.

# Q12: What is the standard normal distribution, and why is it important
# A:
# A normal distribution with mean 0 and standard deviation 1. Important for calculating Z-scores and using statistical tables.

# Q13: What is the Central Limit Theorem (CLT), and why is it critical in statistics
# A:
# The CLT states that the sampling distribution of the mean approaches normality as sample size grows, regardless of population distribution. Critical for inference.

# Q14: How does the Central Limit Theorem relate to the normal distribution
# A:
# It explains why sample means are normally distributed even if the underlying data is not, enabling use of normal-based methods.

# Q15: What is the application of Z statistics in hypothesis testing
# A:
# Z-statistics help test hypotheses about population means when population variance is known, comparing sample mean to hypothesized mean.

# Q16: How do you calculate a Z-score, and what does it represent
# A:
# Z = (X - μ) / σ. It represents how many standard deviations a value is from the mean.

# Q17: What are point estimates and interval estimates in statistics
# A:
# Point estimate: single value estimate of a parameter. Interval estimate: range likely to contain the parameter (e.g., confidence interval).

# Q18: What is the significance of confidence intervals in statistical analysis
# A:
# They provide a range of plausible values for a parameter with a given level of confidence, indicating estimation precision.

# Q19: What is the relationship between a Z-score and a confidence interval
# A:
# Z-scores determine the critical values for confidence intervals. E.g., ±1.96 for 95% confidence.

# Q20: How are Z-scores used to compare different distributions
# A:
# Z-scores standardize values across different scales, making comparisons possible by expressing data in standard deviation units.

# Q21: What are the assumptions for applying the Central Limit Theorem
# A:
# Independent observations, identical distribution, finite mean and variance, and sufficiently large sample size.

# Q22: What is the concept of expected value in a probability distribution
# A:
# The long-run average or mean outcome of a random variable, weighted by probabilities.

# Q23: How does a probability distribution relate to the expected outcome of a random variable
# A:
# The probability distribution describes all possible outcomes and their probabilities, which determine the expected value.

```python
# Q3: Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
# A:
def bernoulli_pdf(x, p):
    return p if x == 1 else (1 - p)
print(bernoulli_pdf(1, 0.6))
print(bernoulli_pdf(0, 0.6))
```

```python
# Q6: Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution
# A:
import numpy as np
import matplotlib.pyplot as plt
values = np.arange(1, 7)
cdf = np.cumsum(np.ones_like(values) / len(values))
plt.step(values, cdf, where='post')
plt.title('CDF of Discrete Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.show()
```

```python
# Q9: Write a Python function to calculate Z-scores from a dataset and plot them
# A:
import numpy as np
import matplotlib.pyplot as plt

def z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    return (data - mean) / std

data = np.random.normal(50, 10, 100)
zs = z_scores(data)

plt.hist(zs, bins=20, edgecolor='black', density=True)
plt.title('Z-score Distribution')
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.show()
```

```python
# Q10: Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
# A:
import numpy as np
import matplotlib.pyplot as plt

population = np.random.exponential(scale=2, size=100000)
sample_means = [np.mean(np.random.choice(population, 30)) for _ in range(1000)]

plt.hist(sample_means, bins=30, edgecolor='black', density=True)
plt.title('Sampling Distribution of the Mean (CLT)')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.show()
```

```python
# Q11: Simulate multiple samples from a normal distribution and verify the Central Limit Theorem
# A:
import numpy as np
import matplotlib.pyplot as plt

sample_means = []
for _ in range(1000):
    sample = np.random.normal(5, 2, 30)
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=30, edgecolor='black', density=True)
plt.title('CLT Verification with Normal Samples')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.show()
```

```python
# Q13: Generate random variables and calculate their corresponding probabilities using the binomial distribution
# A:
import numpy as np
from scipy.stats import binom

n, p = 10, 0.5
x = np.arange(0, n+1)
probs = binom.pmf(x, n, p)
for xi, pi in zip(x, probs):
    print(f"P(X={xi}) = {pi:.4f}")
```

```python
# Q15: Implement hypothesis testing using Z-statistics for a sample dataset
# A:
import numpy as np
from scipy.stats import norm

sample = np.random.normal(50, 10, 30)
pop_mean = 52
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
se = sample_std / np.sqrt(len(sample))
z = (sample_mean - pop_mean) / se
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"Z-statistic: {z:.2f}")
print(f"P-value: {p_value:.4f}")
```

```python
# Q16: Create a confidence interval for a dataset using Python and interpret the result
# A:
import numpy as np
from scipy.stats import norm

data = np.random.normal(100, 15, 50)
mean = np.mean(data)
std = np.std(data, ddof=1)
se = std / np.sqrt(len(data))
z = norm.ppf(0.975)
ci_lower = mean - z * se
ci_upper = mean + z * se

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# Q17: Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
# A:
import numpy as np
from scipy.stats import norm

data = np.random.normal(70, 8, 40)
mean = np.mean(data)
se = np.std(data, ddof=1) / np.sqrt(len(data))
z = norm.ppf(0.975)
ci = (mean - z * se, mean + z * se)

print(f"Sample mean: {mean:.2f}")
print(f"95% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})")
```

```python
# Q21: Write a Python program to compare the standard deviations of two datasets and visualize the difference
# A:
import numpy as np
import matplotlib.pyplot as plt

data1 = np.random.normal(50, 5, 100)
data2 = np.random.normal(50, 15, 100)

std1 = np.std(data1)
std2 = np.std(data2)

print(f"Standard Deviation of Dataset 1: {std1:.2f}")
print(f"Standard Deviation of Dataset 2: {std2:.2f}")

plt.hist(data1, alpha=0.5, label='Dataset 1')
plt.hist(data2, alpha=0.5, label='Dataset 2')
plt.legend()
plt.title('Comparison of Two Datasets')
plt.show()
```

```python
# Q22: Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
# A:
import numpy as np

data = np.random.normal(100, 20, 100)
range_val = np.ptp(data)
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

print(f"Range: {range_val:.2f}")
print(f"IQR: {iqr:.2f}")
```

```python
# Q23: Implement Z-score normalization on a dataset and visualize its transformation
# A:
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(100, 15, 100)
z_normalized = (data - np.mean(data)) / np.std(data)

plt.hist(z_normalized, bins=20, edgecolor='black', density=True)
plt.title('Z-score Normalized Data')
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.show()
```
